```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def alpha_loss(alpha, probability):
    if alpha == 1:
        return -np.log(probability)
    elif alpha == np.inf:
        return 1-probability
    else: return alpha/(alpha-1) * (1-probability**(1-1/alpha))
```

```python
prob_space = np.linspace(1e-6, 1, int(1e6))
alpha_space = [0.25, 0.5, 0.75, 1.0, 2.0, np.inf]
alpha_comments=['','', '','[Log-Loss]','','[0-1 Loss]']
for alpha in alpha_space:
    plt.plot(prob_space, alpha_loss(alpha,prob_space), label=r'$\alpha=${a}'.format(a=alpha))
plt.ylim((0,5))
plt.xlim((0,1))
plt.xlabel('Probability')
plt.ylabel('Loss')
plt.legend()
plt.savefig('loss.eps',dpi=1200)
```
